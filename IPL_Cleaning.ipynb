```python
import pandas as pd

# Load the data
df = pd.read_csv('IPL.csv')

# Show the first 5 rows and the column names
print("✅ Dataset Loaded Successfully!")
print(f"Total Matches in Dataset: {len(df)}")
display(df.head())
```

```python
# 1. See all unique team names currently in the data
if 'batting_team' in df.columns:
    print("Unique Teams before cleaning:")
    print(df['batting_team'].unique())

# 2. Define the changes (Dictionary)
name_changes = {
    'Kings XI Punjab': 'Punjab Kings',
    'Delhi Daredevils': 'Delhi Capitals',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru'
}

# 3. Apply the changes to the whole dataset
df.replace(name_changes, inplace=True)

print("\n✅ Team names standardized!")

# 4. Verify the change
if 'batting_team' in df.columns:
    print("Unique Teams after cleaning:")
    print(df['batting_team'].unique())
```

```python
# 1. Check how many missing (Null) values are in each column
print("Missing values per column:")
print(df.isnull().sum())

# 2. Drop rows where the 'winner' is missing 
# (This removes abandoned/no-result matches)
df.dropna(subset=['winner'], inplace=True)

# 3. Fill other missing values (like 'player_of_match') with "Unknown"
df.fillna('Unknown', inplace=True)

print("\n✅ Missing values handled!")
print(f"Total rows remaining: {len(df)}")
```

```python
print(df.columns.tolist())
```

```python
# Create a 'winner' column from 'match_won_by'
df['winner'] = df['match_won_by']

# Clean up the team names one more time on this specific column
df['winner'].replace(name_changes, inplace=True)

print("✅ 'winner' column created and cleaned!")
```

```python
# Group by 'batter' and sum their runs
top_scorers = df.groupby('batter')['runs_batter'].sum().sort_values(ascending=False)

print("🏏 TOP 10 RUN SCORERS (2008-2025):")
print(top_scorers.head(10))
```

```python
# Sum the 'bowler_wicket' column for each bowler
top_bowlers = df.groupby('bowler')['bowler_wicket'].sum().sort_values(ascending=False)

print("🔥 TOP 10 WICKET TAKERS (2008-2025):")
print(top_bowlers.head(10))
```

```python
# Save the cleaned dataframe
df.to_csv('IPL_2008_2025_Cleaned.csv', index=False)
print("📂 File Saved: 'IPL_2008_2025_Cleaned.csv' is ready for Power BI!")
```
